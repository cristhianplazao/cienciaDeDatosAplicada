## Tabla de contingencias

```python
# Creando dataframe
df=pd.DataFrame({'Sexo':['Mujer','Hombre','Mujer','Hombre','Mujer','Hombre','Mujer','Hombre','Hombre','Mujer','Mujer','Hombre',
                         'Mujer','Hombre','Hombre','Mujer','Mujer','Hombre','Mujer','Hombre','Mujer','Mujer','Hombre','Mujer'
                         ,'Hombre','Mujer','Hombre','Hombre','Mujer',
                         'Mujer','Hombre','Mujer','Hombre','Mujer','Hombre','Mujer','Hombre','Hombre','Mujer','Mujer','Mujer',
                         'Mujer']
                 ,'Ocupacion':['Empleado','Desempleado','Empleado','Desempleado','Empleado','Desempleado','Empleado','Empleado',
                              'Desempleado','Desempleado','Empleado','Desempleado','Empleado','Desempleado','Desempleado',
                               'Desempleado','Empleado','Desempleado','Empleado','Desempleado',
                               'Empleado','Desempleado','Empleado','Empleado',
                              'Desempleado','Desempleado','Empleado','Empleado','Empleado','Empleado','Empleado','Empleado',
                               'Empleado','Empleado','Empleado','Empleado','Empleado','Empleado','Empleado','Desempleado',
                               'Desempleado','Desempleado']})
```

```python
df.head()
```

```python
#### Creación tabla de contingencia

Tabla = pd.crosstab(index=df.Sexo,columns=df["Ocupacion"])
Tabla
```

```python
####Como aparece raro, se va generar la tabla de contingencia de la siguiente forma:


Tabla = pd.crosstab(index=df.Sexo,columns=df["Ocupacion"],margins=True)
Tabla.index= ["Hombre","Mujer","Total_Ocupacion"] # Para las filas, pero el último elemento es el total de la "columna"
Tabla.columns= ["Desempleado", 'Empleado', "Total_Sexo"] # Para las columnas, pero el último elemento es el total de la "fila"

Tabla
```

```python
#### Frecuencia relativa total
Tabla/Tabla.ix['Total_Ocupacion','Total_Sexo']
```

```python
####Frecuencia relativa columnas
Tabla/Tabla.ix["Total_Ocupacion"]
```

```python
####Frecuencia relativa filas
Tabla.div(Tabla.Total_Sexo, axis=0)
```

```python
#### Gráfico que permite relacionar dos variables categóricas
Tabla.plot(kind='bar',figsize=(20,10),legend=True,ylim=(0,50))
```

```python
df.Sexo.value_counts().plot(kind='pie',figsize=(6, 6),title='Sexo',autopct='%.2f')
```

```python
df.Ocupacion.value_counts().plot(kind='pie',figsize=(6, 6),title='Ocupacion',autopct='%.2f')
```

```python
chi2_contingency(Tabla)
```

#### Test de normalidad

```python
df=pd.DataFrame({'normal2':np.random.normal(50,150,5000),'normal50':np.random.normal(150,100,5000),
                 'chi8':np.random.chisquare(8,5000),'chi1':np.random.chisquare(2,5000),
                 'F2':np.random.f(2,50,5000),'F10':np.random.f(10,50,5000)})
```

```python
df.hist(yrot=5.2,grid=False,xlabelsize=15,ylabelsize=15,figsize=(15,15),bins=10) #Ajuste de las clases
```

```python
stats.mstats.normaltest(df['chi8'])
```

```python
stats.mstats.normaltest(df['normal2'])
```

#### H0:Los datos provienen de una distribución normal
